```python
import numpy as np
import PIL as pil
import matplotlib.pyplot as plt
from ciqa import compress, decompress
```

```python
filename = 'images/baboon.bmp'
image = pil.Image.open(filename)
image
```

```python
array = np.asarray(image)
array
```

```python
block_size = 512 // 8
levels = 4
data = compress(image, block_size, levels)
decompressed = decompress(data)
print(decompressed)
pil.Image.fromarray(decompressed)
```
